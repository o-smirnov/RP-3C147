```python
for (p,q),(rr,rl,lr,ll) in ig.items()[:10]:  if type(rr) is not float:     dr,dl = abs(rr-1).mean(),abs(ll-1).mean()    # print IFR, rr and ll offsets, and their reciprocal, for the first 10 baselines    print "%s-%s: %f %f %.2f %.2f"%(p,q,dr,dl,1/dr,1/dl)
```

```python
crl = {}for (p,q),(rr,rl,lr,ll) in ig.iteritems():  if type(rr) is not float:     crl[p,q] = float(abs(rr-1).mean()),float(abs(ll-1).mean())figure(figsize=(16,12))vals = crl.values()plot([cr for cr,cl in vals],[cl for cr,cl in vals],'.w')for pq,(cr,cl) in crl.iteritems():  text(cr,cl,"%s-%s"%pq,horizontalalignment='center',verticalalignment='center',size=8)title("IFR offsets (RR vs. LL)")savefig("IFR-rrll.png",dpi=100);
```
