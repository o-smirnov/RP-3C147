<H1>Chasing after interferometer errors</H1> RP claims his offsets are on the order of 1/400, which suggests the leakage term isresponsible.According to Rick, since IFR response is $V_{rr,pq} = g_p g_q^* (1+d_{rp}d_{lq}^*)I$, and "The 'D' terms are directly related to the ellipticity of the antenna's radiation pattern, which we know is about 5% (circular basis).  So the amplitude of the product is about 1/400, which is 'bang-on' the size of the offsets found above."I say that can't be the whole story, since I'm solving for a full 2x2 Jones matrix -- said solution should incorporate the D terms. So, let's check the size of the offsets I'm finding (_after_ having done a 2x2 solution, so these offsets account for what's left after leakage has beensolved for):

```python
import cPicklefrom Timba.Meq import meq  # needed, otherwise load() complains about MeqVells not knownig = cPickle.load(file('ifrgains.cp'))for (p,q),(rr,rl,lr,ll) in ig.items()[:10]:  if type(rr) is not float:     dr,dl = abs(rr-1).mean(),abs(ll-1).mean()    # print IFR, rr and ll offsets, and their reciprocal, for the first 10 baselines    print "%s-%s: %f %f %.2f %.2f"%(p,q,dr,dl,1/dr,1/dl)
```

...and indeed, bingo, my offsets are typically an order of magnitude lower. So there must be another effect on top of the leakage term.Let's see if there's any systematics here. First, given complex interferometer offsets $c_r$ and $c_l$, let's plot $<|c_r|>$ vs. $<|c_l|>$ (i.e. the average of the absolute value).This is the plot below. At least $c_r$ and $c_l$ correlate nicely. Interestingly, antennas 7 and 14 are very prominent here.

```python
crl = {}for (p,q),(rr,rl,lr,ll) in ig.iteritems():  if type(rr) is not float:     crl[p,q] = float(abs(rr-1).mean()),float(abs(ll-1).mean())figure(figsize=(16,12))plot([cr for cr,cl in crl.itervalues()],[cl for cr,cl in crl.itervalues()],'.w')for pq,(cr,cl) in crl.iteritems():  text(cr,cl,"%s-%s"%pq,horizontalalignment='center',verticalalignment='center',size=8)
```

OK, now let's plot the mean offset per antenna.

```python
ants = set([p for p,q in crl.keys()]+[q for p,q in crl.keys()]);ca = [ (int(p0),[ cr+cl for (p,q),(cr,cl) in crl.iteritems() if p0 in (p,q) ]) for p0 in ants ];ca = [ (p0,sum(x)/(2*len(x))) for p0,x in sorted(ca) ];for p,c in ca:    print "%s: %.2f %f"%(p,1/c,c)## this prints# antenna: 1/c c,          where c is the mean offset across all baselines to that antenna
```

Again, 7 and 14 stick out obviously. Finally, let's plot it against baseline length.

```python
# compute dict of baseline lengthspos = table('3C147_nobl.MS/ANTENNA').getcol('POSITION')name = table('3C147_nobl.MS/ANTENNA').getcol('NAME')baseline = {}for i,p in enumerate(name):    for j,q in enumerate(name):        baseline[p,q] = math.sqrt(sum((pos[i] - pos[j])**2))figure(figsize=(16,12))plot([baseline[pq] for pq,(cr,cl) in crl.iteritems()],[ (cr+cl)/2 for pq,(cr,cl) in crl.iteritems() ],'.w');for pq,(cr,cl) in crl.iteritems():  text(baseline[pq],(cr+cl)/2,"%s-%s"%pq,horizontalalignment='center',verticalalignment='center',size=8)      
```

OK, no real dependence on baseline then -- but highlights antennas 7 and 14 again. FInally, let's plot phase.
